```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_gold = pd.read_csv('E:/UIT/Nam-2/Ki-2/Predicting-Popularity-of-Vietnamese-Articles/2. Data Pre-processing/gold_data.csv')
```

```python
import numpy as np

# 1. Áp dụng log transform trước
df_gold['popularity_score'] = np.log1p(df_gold['popularity_score'] * 10000 + 1)  
# Vẽ biểu đồ phân phối của popularity_score
plt.figure(figsize=(10, 6))
sns.histplot(data=df_gold, x='popularity_score', bins=50)
plt.title('Phân phối của Popularity Score')
plt.xlabel('Popularity Score')
plt.ylabel('Số lượng')
plt.show()

# Vẽ thêm boxplot để thấy rõ outliers
plt.figure(figsize=(10, 4))
sns.boxplot(data=df_gold, x='popularity_score')
plt.title('Boxplot của Popularity Score')
plt.xlabel('Popularity Score')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_gold, x='title_length', y='popularity_score', alpha=0.5)
plt.title('Độ dài tiêu đề vs. Điểm phổ biến')
plt.xlabel('Độ dài tiêu đề')
plt.ylabel('Điểm phổ biến')
plt.tight_layout()

plt.savefig("do_dai_tieu_de.png", dpi=300, bbox_inches='tight')  # dpi cao cho rõ nét
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_gold, x='images', y='popularity_score', alpha=0.5)
plt.title('Số ảnh vs. Điểm phổ biến')
plt.xlabel('Số ảnh')
plt.ylabel('Điểm phổ biến')
plt.tight_layout()

plt.savefig("so_anh.png", dpi=300, bbox_inches='tight')  # dpi cao cho rõ nét
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_gold, x='videos', y='popularity_score', alpha=0.5)
plt.title('Số video vs. Điểm phổ biến')
plt.xlabel('Số video')
plt.ylabel('Điểm phổ biến')
plt.tight_layout()

plt.savefig("so_video.png", dpi=300, bbox_inches='tight')  # dpi cao cho rõ nét
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df_gold, x='posted_hour', y='popularity_score')
plt.title('Giờ đăng bài vs. Điểm phổ biến')
plt.xlabel('Giờ đăng (0–23h)')
plt.ylabel('Điểm phổ biến')
plt.tight_layout()

plt.savefig("gio_dang.png", dpi=300, bbox_inches='tight')  # dpi cao cho rõ nét
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df_gold, x='is_weekday', y='popularity_score')
plt.title('Thứ trong tuần vs. Điểm phổ biến')
plt.xlabel('Thứ (0 = Cuối tuần)')
plt.ylabel('Điểm phổ biến')
plt.tight_layout()

plt.savefig("thu_trong_tuan.png", dpi=300, bbox_inches='tight')  # dpi cao cho rõ nét
plt.show()
```

```python
import ast

df_gold['num_tags'] = df_gold['tags'].apply(
    lambda x: len(ast.literal_eval(x)) if isinstance(x, str) else 0
)
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_gold, x='num_tags', y='popularity_score', alpha=0.5)
plt.title('Số tags vs. Điểm phổ biến')
plt.xlabel('Số lượng tags')
plt.ylabel('Điểm phổ biến')
plt.tight_layout()

plt.savefig("so_tags.png", dpi=300, bbox_inches='tight')  # dpi cao cho rõ nét
plt.show()
```

```python
# Chọn các cột dạng số
numerical_cols = df_gold.select_dtypes(include='number').columns

# Vẽ heatmap tương quan
plt.figure(figsize=(15, 10))
sns.heatmap(df_gold[numerical_cols].corr(), annot=True, cmap='Blues', fmt='.2f')
plt.title('Ma trận tương quan giữa các đặc trưng số')
plt.tight_layout()

plt.savefig("ma_tran_tuong_quan.png", dpi=300, bbox_inches='tight')  # dpi cao cho rõ nét
plt.show()
```
